```python
import pandas as pd
import numpy as np

chiens = pd.DataFrame({
    "nom": ["Rex", "Luna", "Biscuit", "Nougat", "Pixel", "Olive", "Caramel", "Fifi"],
    "race": ["labrador", "jack russell", "bichon", "beagle", "jack russell", "bichon", "labrador", "bichon"],
    "age_annees": [3, 7, 12, 5, 2, 9, 6, 11],
    "poids_kg": [29.5, 6.2, 4.1, np.nan, 5.8, 4.5, 31.0, 3.9],
    "proprio_id": [101, 102, 103, 104, 105, 101, 106, 107]
})

proprios = pd.DataFrame({
    "id": [101, 102, 103, 104, 105, 106, 107],
    "ville": ["Bruxelles", "Liège", "Forest", "Namur", "Bruxelles", "Gand", "Uccle"]
})
```

1. Affiche les 3 premières lignes et les dimensions du DataFrame.

```python
chiens.info()
```

```python
display(chiens.head(3))
display(chiens.shape)
display(proprios.head(3))
display(proprios.shape)
```

2. Affiche le nom et l'âge du chien en position 4 (par position, pas par label).

```python
display(chiens.iloc[4])
display(chiens[['nom','age_annees']].iloc[[4]])
```

3. Affiche tous les chiens de plus de 5 kg.

```python
display(chiens.loc[chiens['poids_kg']>5])
```

4. Affiche les jack russell de moins de 8 ans.

```python
display(chiens.loc[(chiens['race']=='jack russell')&(chiens['age_annees']<8)])
```

5. Crée une colonne age_humain (âge × 7).

```python
chiens['age_humain']=chiens['age_annees']*7
display(chiens)
```

6. Quels sont les 3 chiens les plus lourds ?

```python
display(chiens.sort_values('poids_kg',ascending=False).head(3))
```

7. Combien y a-t-il de chiens par race ?

```python
display(chiens['race'].value_counts())
```

8. Quel est le poids moyen par race ?

```python
display(chiens.groupby('race')['poids_kg'].mean())
```

9. En une seule opération : poids moyen ET âge maximum par race, avec des colonnes proprement nommées.

```python
display(chiens.groupby('race').agg(poids_moyen=('poids_kg','mean'),age_max=('age_annees','max')))
```

10. Un poids manque dans les données. Trouve-le, puis remplace-le par le poids moyen.

```python
display(chiens[chiens['poids_kg'].isna()])
chiens['poids_kg']=chiens['poids_kg'].fillna(chiens['poids_kg'].mean())
display(chiens)
```

11. Affiche chaque chien avec la ville de son propriétaire (chiens.proprio_id ↔ proprios.id, noms différents).

```python
display(chiens.merge(proprios, left_on="proprio_id", right_on="id", how="left"))
```

12. Dans quelle ville vit le chien le plus âgé ? Une seule chaîne d'opérations.

```python
display(chiens.sort_values("age_annees", ascending=False).head(1).merge(proprios, left_on="proprio_id", right_on="id", how="left"))
```

Autre option plus solide (merger d'abord comme ça tout le monde à sa ville puis isoler le top 1 âge puis demander d'afficher la ville seulement)

```python
display(chiens.merge(proprios, left_on="proprio_id", right_on="id", how="left").sort_values("age_annees", ascending=False).head(1)["ville"])
```
