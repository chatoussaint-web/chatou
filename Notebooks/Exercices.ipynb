```python
def inventaire(stock, seuil):
    print(f"Produits en dessous du seuil de {seuil} :")
    resultat = {produit: quantite for produit, quantite in stock.items() if quantite < seuil}
    for produit, quantite in resultat.items():
        print(f"{produit} : {quantite}")
    return len(resultat)

stock = {
    "pommes": 30,
    "bananes": 5,
    "cerises": 2,
    "mangues": 15,
    "fraises": 1
}
print(inventaire(stock, 10))
```

```python
def rapports_eleves(eleves):
    for eleve in eleves:
        nom = eleve["nom"]
        note = eleve["note"]
        age = eleve["age"]
        if note < 10:
            print(f"{nom} : {age} ans : {note}/20 : Échec")
        elif 10 <= note <= 13:
            print(f"{nom} : {age} ans : {note}/20 : Suffisant")
        elif 14 <= note <= 16:
            print(f"{nom} : {age} ans : {note}/20 : Bien")
        else:
            print(f"{nom} : {age} ans : {note}/20 : Très bien")
    moyenne_classe = sum(eleve["note"] for eleve in eleves) / len(eleves)
    moyenne_age = sum(eleve["age"] for eleve in eleves) / len(eleves)
    return moyenne_classe, moyenne_age

eleves = [
    {"nom": "Alice", "note": 17, "age": 20},
    {"nom": "Bob", "note": 9, "age": 19},
    {"nom": "Clara", "note": 13, "age": 21},
    {"nom": "Driss", "note": 11, "age": 18},
    {"nom": "Eva", "note": 19, "age": 22}
]

moyenne_classe, moyenne_age = rapports_eleves(eleves)
print(f"Moyenne de la classe : {moyenne_classe}/20", "et", f"moyenne d'âge : {moyenne_age} ans")
```

```python
def analyse_texte(texte):
    mots = texte.split()
    nombre_mots = len(mots)
    nombre_caracteres = len(texte.replace(" ", ""))
    mot_le_plus_long = max(mots, key=len)
    unique = set(mots)
    nombre_mots_uniques = len(unique)
    return {
        "nombre_mots": nombre_mots,
        "nombre_caracteres": nombre_caracteres,
        "mot_le_plus_long": mot_le_plus_long,
        "nombre_mots_uniques": nombre_mots_uniques
    }

texte = "le chat mange le poisson et le chien mange la viande"
print(analyse_texte(texte))
```

2. Compter les voyelles
Écris une fonction compter_voyelles(chaine) qui retourne le nombre de voyelles (a, e, i, o, u, y — sans accents pour l'instant) dans une chaîne, insensible à la casse.
pythoncompter_voyelles("Bonjour")   # → 3
compter_voyelles("Python")    # → 1
compter_voyelles("AEIOU")     # → 5

```python
def compter_voyelles(mot):
    voy = {'a','e','i','o','u','y'}
    mot2 = mot.lower()
    compteur=0
    for i in (mot2):
        if i in voy:
            compteur=compteur+1
    return compteur

display(compter_voyelles('Bonjour'))
display(compter_voyelles('Python'))
display(compter_voyelles('AEIOU'))
```

3. Palindrome
Écris une fonction est_palindrome(chaine) qui vérifie si une chaîne se lit pareil dans les deux sens, en ignorant la casse et les espaces.
palindrome("kayak")          # → True
palindrome("Engage le jeu que je le gagne")  # → True
palindrome("python")         # → False
Essaie sans utiliser le slicing [::-1] cette fois — boucle ou récursion.

```python
def palindrome(chaine):
    chaine2=(chaine).replace(' ','').lower()
    if chaine2==chaine2[::-1]:
        return True
    else: return False

display(palindrome("kayak"))
display(palindrome("Engage le jeu que je le gagne"))
display(palindrome("python"))
```

```python
#Autre méthode (sans slicing)
def palindrome_algo(chaine):
    chaine2 = chaine.replace(' ', '').lower()
    gauche, droite = 0, len(chaine2) - 1
    while gauche < droite:
        if chaine2[gauche] != chaine2[droite]:
            return False
        gauche += 1
        droite -= 1
    return True

display(palindrome_algo("kayak"))
display(palindrome_algo("Engage le jeu que je le gagne"))
display(palindrome_algo("python"))
```

4. FizzBuzz
Le classique des entretiens techniques. Écris une fonction fizzbuzz(n) qui parcourt les nombres de 1 à n et :

affiche "Fizz" si le nombre est divisible par 3
affiche "Buzz" si le nombre est divisible par 5
affiche "FizzBuzz" si le nombre est divisible par 3 ET par 5
affiche le nombre lui-même sinon

```python
def fizzbuzz(n):
    for i in range(1,n+1):
        if i%3==0 and i%5!=0:
            print('Fizz')
        elif i%5==0 and i%3!=0:
            print('Buzz')
        elif i%3==0 and i%5==0:
            print('FizzBuzz')
        else:
            print(i)

fizzbuzz(15)
```

5. Nombre premier
Écris une fonction est_premier(n) qui retourne True si n est un nombre premier, False sinon.
Rappel : un nombre premier n'est divisible que par 1 et par lui-même, et doit être supérieur à 1.
pythonest_premier(7)   # → True
est_premier(10)  # → False
est_premier(1)   # → False
est_premier(2)   # → True
Indice si tu bloques : tu n'as besoin de tester les diviseurs que jusqu'à la racine carrée de n (mais tu peux aussi juste tester jusqu'à n-1 pour commencer, plus simple).

```python
def premier(n):
    if n>1:
        for diviseur in range(2,n-1):
            if n%(diviseur)==0:
                return False
        return True
    return False
display(premier(9))
display(premier(7))
display(premier(10))
display(premier(1))
display(premier(2))
```

6. Somme des chiffres
Écris une fonction somme_chiffres(n) qui retourne la somme des chiffres d'un nombre entier positif.
somme_chiffres(1234)   # → 10  (1+2+3+4)
somme_chiffres(7)      # → 7
somme_chiffres(905)    # → 14  (9+0+5)
Essaie sans convertir n en chaîne de caractères (str()) — uniquement avec des opérations arithmétiques (% et //).

```python
def somme(n):
    total=0
    while n!=0:
        total+=n%10
        n=n//10
    return total
    
display(somme(1234))
display(somme(7))
display(somme(905))
```

7. Liste sans doublons
Écris une fonction sans_doublons(lst) qui retourne une nouvelle liste contenant les éléments de lst, mais sans les doublons — en gardant l'ordre d'apparition d'origine.
sans_doublons([1, 2, 2, 3, 1, 4])        # → [1, 2, 3, 4]
sans_doublons(['a', 'b', 'a', 'c', 'b']) # → ['a', 'b', 'c']
Contrainte : sans utiliser set() directement sur la liste entière (tu peux quand même t'en servir comme outil auxiliaire si tu veux, mais essaie d'abord avec juste une boucle et un test d'appartenance in).

```python
def sans_doublons(lst):
    resultat = []
    for element in lst:
        if element not in resultat:
            resultat.append(element)
    return resultat
                
display(sans_doublons([1, 2, 2, 3, 1, 4]))
display(sans_doublons(['a', 'b', 'a', 'c', 'b']))        
```

8. Recherche du maximum
Écris une fonction mon_max(lst) qui retourne la plus grande valeur d'une liste de nombres — sans utiliser la fonction native max().
mon_max([3, 7, 2, 9, 4])    # → 9
mon_max([-5, -1, -10])      # → -1
mon_max([42])               # → 42
Réfléchis à la stratégie avant de coder : qu'est-ce que tu dois garder en mémoire au fur et à mesure que tu parcours la liste ?

```python
def mon_max(lst):
    resultat=lst[0]
    for el in lst:
        if el>resultat:
           resultat=el
    return resultat

display(mon_max([3, 7, 2, 9, 4]))
display(mon_max([-5, -1, -10]))
display(mon_max([42]))
```

9. Compter les occurrences
Écris une fonction compter_occurrences(lst, valeur) qui retourne le nombre de fois qu'une valeur apparaît dans une liste — sans utiliser la méthode .count().
compter_occ([1, 2, 3, 2, 2, 4], 2)   # → 3
compter_occ(['a', 'b', 'a'], 'c')    # → 0
compter_occ([5, 5, 5], 5)            # → 3

```python
def compter_occ(lst,valeur):
    resultat=0
    for el in lst:
        if el==valeur:
            resultat+=1
    return resultat

display(compter_occ([1, 2, 3, 2, 2, 4], 2))
display(compter_occ(['a', 'b', 'a'], 'c'))
display(compter_occ([5, 5, 5], 5))
```

10. Inverser une liste
Écris une fonction mon_inverse(lst) qui retourne une nouvelle liste avec les éléments dans l'ordre inverse — sans utiliser [::-1] ni .reverse() ni reversed().
mon_inverse([1, 2, 3, 4])        # → [4, 3, 2, 1]
mon_inverse(['a', 'b', 'c'])     # → ['c', 'b', 'a']
mon_inverse([1])                 # → [1]
Tu as déjà résolu un problème très proche de celui-ci, dans un exercice précédent — à quoi ça te fait penser ?

```python
def mon_inverse(lst):
    resultat=[]
    for el in lst:
        resultat=[el] + resultat
    return resultat

display(mon_inverse([1, 2, 3, 4]))
display(mon_inverse(['a', 'b', 'c']))
display(mon_inverse([1]))   
```
