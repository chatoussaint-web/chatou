Exercice 1 : Manipulation de chaîne

Tu as une phrase en entrée, par exemple : "Bonjour le monde"

Écris une fonction qui :

    Compte combien de fois chaque mot apparaît dans la phrase
    Retourne un dictionnaire avec le mot comme clé et le comptage comme valeur

```python
def wordcount(phrase):
    lst=phrase.lower().split()
    dct={}
    for word in lst:
        if word in dct:
             dct[word]+=1
        else: dct[word]=1
    return dct

wordcount('Bonjour le monde le monde est beau')       
```

Exercice 2 : Un cran au-dessus

Tu as une liste de nombres : [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

Écris une fonction qui retourne :

    Le nombre le plus fréquent
    Combien de fois il apparaît

Si deux nombres ont la même fréquence, retourne n'importe lequel des deux.

```python
def numcount(num):
    dct={}
    for el in num:
        if el in dct:
             dct[el]+=1
        else: dct[el]=1
    plusfrqt=max(dct,key=dct.get)
    print(f'le nombre {plusfrqt} a pour fréquence {dct[plusfrqt]}')
    return (plusfrqt,dct[plusfrqt])
    

numcount([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5])       
```

Exercice 3 : Manipulation de liste

Tu as une liste d'intérêts : [3, 1, 4, 1, 5, 9, 2, 6]

Écris une fonction qui retourne une nouvelle liste contenant uniquement les éléments pairs, triés par ordre croissant.

Attendu pour cet exemple : [2, 4, 6]

```python
def lstpairs(lst):
    lst2=[]
    for el in lst:
        if el%2==0:
            lst2.append(el)
        else:
            continue
    return sorted(lst2)

lstpairs([3, 1, 4, 1, 5, 9, 2, 6])
```

Exercice 4 : Un cran supplémentaire

Tu as une liste de phrases :

phrases = ["Bonjour le monde", "la vie est belle", "le monde tourne"]

Écris une fonction qui :

    Combine toutes les phrases en un seul texte
    Compte combien de fois chaque mot apparaît
    Retourne les trois mots les plus fréquents avec leur fréquence

```python
def combi(phrases):
    dct = {}
    for phrase in phrases:
        for word in phrase.split():  # sépare la phrase en mots
            if word in dct:
                dct[word] += 1
            else:
                dct[word] = 1
    dct_sort=sorted(dct, key=dct.get, reverse=True)[:3]
    resultat=[]
    for word in dct_sort:
        resultat.append((word, dct[word]))
    return dct, dct_sort, resultat

combi(["Bonjour le monde", "la vie est belle", "le beau monde tourne", "le monde est beau"])       
```

Exercice 5 : Dictionnaires imbriqués

etudiants = [
    {"nom": "Alice", "notes": [12, 15, 18]},
    {"nom": "Bob", "notes": [10, 11, 9]},
    {"nom": "Charlie", "notes": [14, 16, 17]}
]

Écris une fonction qui :

    Calcule la moyenne de chaque étudiant
    Retourne le nom de l'étudiant avec la meilleure moyenne

```python
def moystd(dcts):
    dct2={}
    for std in dcts:
        nom=std.get('nom')
        notes=std.get('notes')
        moy=sum(notes)/len(notes)
        dct2[nom]=moy
    maximum=max(dct2,key=dct2.get)
    return maximum


moystd([
    {"nom": "Alice", "notes": [12, 15, 18]},
    {"nom": "Bob", "notes": [10, 11, 9]},
    {"nom": "Charlie", "notes": [14, 16, 17]},
    {'nom':'Charlotte','notes':[13,12,16]}
])      
```

Exercice 6 : Filtrer selon un seuil

Écris une fonction qui retourne la liste des noms des étudiants dont la moyenne est supérieure ou égale à 13.

Attendu : ["Alice", "Charlie", "Charlotte"]

```python
def fltrstd(dcts):
    lst=[]
    for std in dcts:
        nom=std.get('nom')
        notes=std.get('notes')
        moy=sum(notes)/len(notes)
        if moy>=13:
           lst.append(nom)
    return lst


fltrstd([
    {"nom": "Alice", "notes": [12, 15, 18]},
    {"nom": "Bob", "notes": [10, 11, 9]},
    {"nom": "Charlie", "notes": [14, 16, 17]},
    {'nom':'Charlotte','notes':[13,12,16]}
])      
```
